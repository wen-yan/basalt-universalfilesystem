```
#r "nuget:AWSSDK.S3,3.7.412.5"
```

```
using Amazon;
using Amazon.Runtime;
using Amazon.Runtime.CredentialManagement;
using Amazon.S3;
using Amazon.S3.Model;
using System.IO;
```

```
AWSCredentials credentials = new BasicAWSCredentials("test", "test");
AmazonS3Config config = new AmazonS3Config()
{
    ServiceURL = "http://localhost:4566",
    ForcePathStyle = true,
};

IAmazonS3 client =  new AmazonS3Client(credentials, config);
```

```
// create bucket
var request = new PutBucketRequest
{
    BucketName = "ufs-test-bucket",
};

await client.PutBucketAsync(request)
```

```
// list buckets
await client.ListBucketsAsync()
```

```
// add items
string[] paths = new[]
{
    "dir1/file1.txt",
    "dir1/dir2/file2.txt",
    "dir1/dir3/file3.txt",
    "dir1/dir3/dir4/file4.txt",
    "dir5/file5.txt",
    "dir5/dir6/file6.txt",
    
};

foreach(string path in paths)
{
    MemoryStream stream = new MemoryStream();
    stream.WriteByte(1);
    PutObjectRequest request = new()
    {
        BucketName = "ufs-test-bucket",
        Key = path,
        InputStream = stream,
    };
    await client.PutObjectAsync(request, default);
}
```

```
ListObjectsV2Request request = new()
{
    BucketName = "ufs-test-bucket",
    Prefix = "dir",
    Delimiter = "/",
};

await client.ListObjectsV2Async(request, default)
```

```
// get metadata from prefix
GetObjectMetadataRequest request = new()
{
    BucketName = "ufs-test-bucket",
    Key = "dir1/",
};

Exception e;
try
{
    await client.GetObjectMetadataAsync(request, default);
}
catch(Exception ex)
{
    e = ex;
}
e
```
