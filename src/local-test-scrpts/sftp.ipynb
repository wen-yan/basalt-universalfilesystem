```polyglot-notebook
using Renci.SshNet;
using Renci.SshNet.Async;
using System.IO;
using System.Threading;

string host = "localhost";
int port = 2222;
string username = "demo";
string password = "demo";

SftpClient client = new SftpClient(host, port, username, password);

client.Connect();
```

```polyglot-notebook
// add items
void EnsureDirectoryExists(SftpClient client, string path)
{
    if (string.IsNullOrEmpty(path))
        return;

    if (!client.Exists(path))
    {
        // Recursively create parent first
        string parent = Path.GetDirectoryName(path).Replace("\\", "/");
        EnsureDirectoryExists(client, parent);

        client.CreateDirectory(path);
    }
}


string[] paths = new[]
{
    "/sftp/dir1/file1.txt",
    "/sftp/dir1/dir2/file2.txt",
    "/sftp/dir1/dir3/file3.txt",
    "/sftp/dir1/dir3/dir4/file4.txt",
    "/sftp/dir5/file5.txt",
    "/sftp/dir5/dir6/file6.txt",
    
};

foreach(string path in paths)
{
    MemoryStream stream = new MemoryStream();
    stream.WriteByte(1);
    stream.Position = 0;

    string directory = Path.GetDirectoryName(path).Replace("\\", "/");
    EnsureDirectoryExists(client, directory);

    await client.UploadAsync(stream, path, true);
}
```
